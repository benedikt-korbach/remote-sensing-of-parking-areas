```python
# Adjust these paths to where your datasets are stored
image_folder = "03_create_masks/train_validate/images"
mask_folder = "03_create_masks/train_validate/masks"

# Load images and masks
def load_dataset(image_folder, mask_folder):
    images = []
    masks = []
    for img_name in os.listdir(image_folder):

        if not img_name.lower().endswith(".png"):
            continue

        img_path = os.path.join(image_folder, img_name)
        mask_path = os.path.join(mask_folder, img_name)  # Assuming same names for images and masks

        img = cv2.imread(img_path)
        mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)


        print(f"Reading: {img_path}, Shape: {img.shape if img is not None else 'None'}")  # Debugging line
        print(f"Reading: {mask_path}, Shape: {mask.shape if mask is not None else 'None'}")  # Debugging line


        images.append(img_to_array(img))
        masks.append(img_to_array(mask))

    images = np.array(images, dtype="float32")
    masks = np.array(masks, dtype="float32")
    masks /= 255.0  # Normalize masks to [0, 1]

    # Convert masks to categorical (one-hot encoding)
    masks = to_categorical(masks, num_classes=3)  # Background, Car, Truck
    return images, masks

images, masks = load_dataset(image_folder, mask_folder)

# Split dataset
x_train, x_val, y_train, y_val = train_test_split(images, masks, test_size=0.2, random_state=42)
```
