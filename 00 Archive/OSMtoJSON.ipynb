```python
import pandas as pd
import json

# Path to your JSON file
json_file_path = '/content/drive/MyDrive/Master Thesis/01 Data Acquisition/01b Truck Parking GeoJSON/netherlands_truck_parking_100_motorway.geojson'

# Load JSON data from the file
with open(json_file_path, 'r') as file:
    json_data = json.load(file)

# Check if the data is under a key (like 'features') or directly an array of features
truck_parking = json_data['features']
truck_parking[0]
```

```python
# Initialize list to hold extracted data
truck_parking_csv = []

# Extract relevant data from each item
for item in truck_parking:
    osm_id = item["properties"].get("@id")
    osm_way_id = osm_id.split("/")[-1]
    geometry = "POLYGON((" + str(item["geometry"]["coordinates"]).replace('[','').replace(']','') + "))"

    # Flatten all_properties into a single string
    all_properties = ", ".join(f'{k}="{v}"' for k, v in item["properties"].items())

    row = {
        "osm_version": item["properties"].get("@version"),
        "osm_way_id": osm_way_id,
        "name": item["properties"].get("name"),
        "osm_timestamp": item["properties"].get("@timestamp"),
        "geometry": geometry,
        "all_tags": all_properties,
        "hgv": item["properties"].get("hgv"),
        "parking": item["properties"].get("parking"),
        "capacity": item["properties"].get("capacity"),
        "amenity": item["properties"].get("amenity")
    }
    truck_parking_csv.append(row)

# Create DataFrame
truck_parking_df = pd.DataFrame(truck_parking_csv)
truck_parking_df
```

```python
# Save the DataFrame as a CSV file
truck_parking_df.to_csv('/content/drive/MyDrive/Master Thesis/01 Data Acquisition/01b Truck Parking GeoJSON/netherlands_truck_parking_100_motorway_new_csv', index=False)
```

```python
# Convert the DataFrame to JSON and save it to the specified path
truck_parking_df.to_json('/content/drive/MyDrive/Master Thesis/01 Data Acquisition/01b Truck Parking GeoJSON/netherlands_truck_parking_100_motorway_.geojson', orient='records')

# Print a confirmation message
print(f'DataFrame has been saved as {json_file_path}')
```
